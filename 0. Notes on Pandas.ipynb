Reading in data from a CSV utilizing Pandas:
----

```python
# Create a Pandas dataframe of the SeattleWeather CSV.
df = pd.read_csv("SeattleWeather_1948-2017.csv")

# Fill empty values with ("")
df = df.fillna("")

# Print the type and the dimensions of the dataframe
print("var df Data type: " + str(type(df)))
print("var df Shape    : " + str(df.shape))

# Print a summary of the first 10 head values of the data
df.head(10)
```

```python
# Print a summary of the first 10 tail values of the data
df.tail(10)
```

```python
# Printing a certain column's values by name; Printing the first 5 dates.
# Note, utilizing head() and tail() also works.
df["DATE"][:5]
```

```python
# Printing the first five rows
df[:5]
```

```python
# Selecting a specific row
# ix(), loc(), iloc() are all the same command.
df.iloc[3]
```
