```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%matplotlib inline


x  = [1, 2, 4, 8, 12]
s1 = [29807002, 18413167, 17631680, 13238196, 12815418] 
s2 = [1924514, 2176668, 3429564, 6129782, 6354216]

xaug = np.arange(1, 13, 1)
funct_s1 = interp1d(x, s1, kind="linear")
funct_s2 = interp1d(x, s2, kind="linear")
y_s1 = []
y_s2 = []

for x in range(len(xaug)):
    y_s1.append(funct_s1(xaug[x]))
    y_s2.append(funct_s2(xaug[x]))
    

plt.plot(xaug, y_s1)
plt.ylabel("S1 in millions")
plt.xlabel("Number of cores")
plt.title("S1 linear function")
```

```python
plt.plot(xaug, y_s2)
plt.ylabel("S2")
plt.xlabel("Number of cores")
plt.title("S2 linear function")
```
