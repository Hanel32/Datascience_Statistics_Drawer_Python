```python
import math
import numpy as np
import matplotlib.pyplot as pp

# Allows matplotlib to be rendered inline:
%matplotlib inline

# Junky comparison between trigonometric functions:
x = np.linspace(-10, 10, 100)
y = np.sin(x)
z = np.cos(x)
p = np.tan(x)
pp.plot(x, y)
pp.plot(x, z)
pp.plot(x, p)
pp.title("Comparison of sin(), cos(), tan():")
pp.legend(["sin", "cos", "tan"])
```
